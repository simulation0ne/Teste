# 1 - Criando a função de Aptidão

Utilizando Jupyter e NumPy: para a função de variável contínua apresentada, construa uma função para o cálculo do “custo” (função-custo).

Assinatura: **def fitness(individual)**


<center>
    $
        \large
         f(x,y) = 0.5 - \frac{ \left( \sin\sqrt{x^2+y^2} \right)^2 - 0.5} { \left(1+0.001\cdot(x^2+y^2) \right)^2}
    $
</center>


fitness (individual)


x = individual [0]

y = individual [1]

np.sin()


```python
#CÉLULA PSO-LIB-01
import numpy as np
import math
import matplotlib.pyplot as plt
from enum import Enum
from operator import xor
%matplotlib inline
```

```python
#CÉLULA PSO-LIB-02
# Função para cálculo de fitness (aptidão de cada individuo)
def fitness(individual) :  
    
    x = individual [0]
    y = individual [1]
    
    num = np.sin(np.sqrt(x**2+y**2))**2 - 0.5
    den = (1+0.001*(x**2+y**2))**2
    
    z = 0.5 - (num/den)
    return z

    
```
